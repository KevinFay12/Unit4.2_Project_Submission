```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots in-notebook
%matplotlib inline

# Increase default figure and font sizes for easier viewing.
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14

plt.style.use('fivethirtyeight')
```

```python
path = 'data/loan.csv'
loan = pd.read_csv(path, low_memory=False)
```

```python
loan.head()
```

```python
loan.shape
```

```python
loan.isnull().sum()
```

```python
half_count = len(loan) / 2
loan = loan.dropna(thresh=half_count,axis=1)
loan.shape
#dropping columsnw ith more than half of values missing
```

```python
loan.describe()
```

```python
loan.drop(['id','member_id', 'url', ],1, inplace=True)
print(loan.columns)
#no reason to have id, member_id, url at this time
```

```python
loan_int_rate = loan.int_rate.values
loan.drop('int_rate',axis = 1)
loan_int_rate
```

```python
fig, ax = plt.subplots(figsize=(14,7))
pd.Series(loan_int_rate).plot(kind='hist',alpha=.7, bins=20, title='Interest Rate Distribution', ax=ax)
```

```python
loan = loan[(loan.loc[:,'loan_status'] == 'Fully Paid') | (loan.loc[:, 'loan_status'] == 'Charged Off')]

status_dict = {'loan_status': { "Fully Paid": 1, "Charged Off": 0}}
loan = loan.replace(status_dict)
```

```python
status_amt = loan.loc[:,'loan_status'].value_counts()
print(str(status_amt))

fig, ax = plt.subplots(figsize=(14,7))

status_amt.plot(kind='bar', ax=ax)
```

```python
#all correlated data
cor = loan.corr()
print(cor)
```

```python
from sklearn.linear_model import LinearRegression
X = loan.loc[:, ['loan_status']]
y = loan.loc[:, 'dti']

lr = LinearRegression()

lr.fit(X,y)
```

```python
# 4. Use the model to make predictions.
# We will add the fitted values to our DataFrame.
# /scrub/
loan.loc[:, 'predictions'] = lr.predict(X)
print(lr.predict(X))
```

```python
fig, ax = plt.subplots()
loan.plot(kind='scatter', x='dti', y='loan_status', ax=ax, alpha=.2)
ax.plot([0, 40], [0, 1], 'r');
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

feature_col = ['dti']

X = loan.loc[:, feature_col]
y = loan.loc[:, 'loan_status']

logreg = LogisticRegression()

logreg.fit(X,y)
pred =logreg.predict(X)
```

```python
fig, ax = plt.subplots()
loan.plot(kind='scatter', x='dti', y='loan_status', ax=ax)
ax.plot(loan.loc[:, 'dti'], np.array(pred), c='r');
```
